```python
import tensorflow as tf
import numpy as np
import pandas as pd
```

```python
zoo_df = pd.read_csv("testdata/zoo.csv", header=None)
```

```python
zoo_df.head()
```

```python
zoo_df.columns=["hair"     ,
"feathers"  ,
"eggs"      ,
"milk"     ,
"airborne"  ,
"aquatic" ,
"predator"  ,
"toothed"   ,
"backbone"  ,
"breathes"  ,
"venomous"  ,
"fins"      ,
"legs"      ,
"tail"      ,
"domestic"  ,
"catsize"   ,
"type"      ]
```

```python
zoo_df.head()
```

```python
zoo_x = zoo_df[["hair"     ,
"feathers"  ,
"eggs"      ,
"milk"     ,
"airborne"  ,
"aquatic" ,
"predator"  ,
"toothed"   ,
"backbone"  ,
"breathes"  ,
"venomous"  ,
"fins"      ,
"legs"      ,
"tail"      ,
"domestic"  ,
"catsize"   ]]
```

```python
zoo_y = zoo_df['type']
zoo_y.head()
```

```python
zoo_y = np.array(zoo_y).reshape(len(zoo_y),1)
```

```python
zoo_y.shape
```

```python
# x = tf.placeholder(tf.float32, [None, 16])
# y = tf.placeholder(tf.int32, [None, 1])

x = tf.placeholder(tf.float32, [None, 16])
y = tf.placeholder(tf.int32, [None, 1])

y_one_hot = tf.one_hot(y, 7)
y_one_hot = tf.reshape(y_one_hot, [-1, 7]) # nx7
```

```python
tf.set_random_seed(123)
w = tf.Variable(tf.random_normal([16, 7]))
b = tf.Variable(tf.random_normal([7]))

logit = tf.matmul(x,w) + b
model = tf.nn.softmax(logit)
cost_i = tf.nn.softmax_cross_entropy_with_logits(logits=logit, labels=y_one_hot)
cost = tf.reduce_mean(cost_i)

optimizer = tf.train.AdamOptimizer(0.1).minimize(cost)

predict = tf.argmax(model, 1)
correct_pred = tf.equal(predict, tf.argmax(y_one_hot, 1))
accuracy = tf.reduce_mean(tf.cast(correct_pred, tf.float32))
```

```python
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    
    for s in range(3000):
        sess.run(optimizer, feed_dict={x:zoo_x, y:zoo_y})
        if s % 100 == 0:
            loss, acc = sess.run([cost, accuracy], feed_dict = {x:zoo_x, y:zoo_y})
            
            print("s:{}, cost:{}, acc:{}".format(s, loss, acc))
```
