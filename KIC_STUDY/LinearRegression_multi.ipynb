<b>1. 학습 데이터를 벡터 형태로 만들어서 회귀분석 수행</b>

```python
import tensorflow as tf
```

```python
x1 = [1,0,3,0,6]
x2 = [0,2,0,4,5]
y = [1,2,3,4,5]

w1 = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
w2 = tf.Variable(tf.random_uniform([1], -1.0, 1.0))
b =tf.Variable(tf.random_uniform([1], -1.0, 1.0))
```

```python
h = x1*w1 + x2*w2 + b
cost = tf.reduce_mean(tf.square(y-h))
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(1000):
        _, c_w1, c_w2, c_b, c_cost = sess.run([train, w1, w2 , b, cost])
        if step % 100 == 0:
            print(c_w1, c_w2, c_b, c_cost)
    print(sess.run(h))
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
```

```python
x1 = tf.placeholder(tf.float32)
x2 = tf.placeholder(tf.float32)
x3 = tf.placeholder(tf.float32)

y = tf.placeholder(tf.float32)

w1 = tf.Variable(tf.random_normal([1]))
w2 = tf.Variable(tf.random_normal([1]))
w3 = tf.Variable(tf.random_normal([1]))
b = tf.Variable(tf.random_normal([1]))

h = x1*w1 + x2*w2 + x3*w3 + b

cost = tf.reduce_mean(tf.square(h-y))
optimizer = tf.train.GradientDescentOptimizer(0.00002)
train = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())

    for step in range(100):
        _, c_w1, c_w2, c_w3, c_b, c_cost, c_h = sess.run([train, w1, w2 , w3, b, cost, h],
                                              feed_dict ={x1:x1_data, x2:x2_data, x3:x3_data, y:y_data})
        if step % 10000 == 0:
            print(c_w1, c_w2, c_w3, c_b, c_cost)
            
    print(sess.run(h, feed_dict={x1:x1_data, x2:x2_data, x3:x3_data}))
```

<b>2. 위에서 실습한 데이터를 행렬로 바꾸어 실행해본다.</b>

```python
import numpy as np
```

```python
x1_data = [73., 93., 89., 96., 73.]
x2_data = [80., 88., 91., 98., 66.]
x3_data = [75., 93., 90., 100., 70.]
y_data = [152., 185., 180., 196., 142.]
```

```python
x_data = np.array([np.array(x1_data), np.array(x2_data), np.array(x3_data)])
x_data = x_data.T
```

```python
x_data
```

```python
y_data = np.array(y_data).reshape(5,1)
```

```python
y_data
```

```python
x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])

w = tf.Variable(tf.random_normal([3,1]))
b = tf.Variable(tf.random_normal([1]))

h = tf.matmul(x, w) + b

cost  = tf.reduce_mean(tf.square(h-y))
optimizer = tf.train.GradientDescentOptimizer(0.00002)
train = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10000):        
        _, cur_w, cur_cost = sess.run([train, w, cost], feed_dict = {x:x_data, y:y_data})
        
        if step % 1000 ==0:
            print("w: \n", cur_w,"\n cost:",  cur_cost)
            print()
            
    print(sess.run(h, feed_dict={x:x_data}))
```

<b>3. 새로운 데이터로 동일하게 회귀분석을 실행 ( 행렬의 차원은 동일 ) </b>

```python
exam_data = np.array([73,80,75,152,
                    93,88,93,185,
                    89,91,90,180,
                    96,98,100,196,
                    73,66,70,142,
                    53,46,55,101]).reshape(6,4)
```

```python
y_data
```

```python
x_data
```

```python
x = tf.placeholder(tf.float32, shape=[None, 3])
y = tf.placeholder(tf.float32, shape=[None, 1])

w = tf.Variable(tf.random_normal([3,1]))
b = tf.Variable(tf.random_normal([1]))

h = tf.matmul(x, w) + b

cost  = tf.reduce_mean(tf.square(h-y))
optimizer = tf.train.GradientDescentOptimizer(0.00002)
train = optimizer.minimize(cost)

with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for step in range(10000):        
        _, cur_w, cur_cost = sess.run([train, w, cost], feed_dict = {x:x_data, y:y_data})
        
        if step % 1000 ==0:
            print("w: \n", cur_w,"\n cost:",  cur_cost)
            print()
            
    print(sess.run(h, feed_dict={x:x_data}))
```
